### t2.micro, AWS Marketplace -> Anaconda with Python 3

### 1) Heat Map

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston

boston = load_boston()
df     = pd.DataFrame(boston.data, columns=boston.feature_names)

plt.figure(figsize=(20,4))
sns.heatmap(df.corr(), annot=True, cmap='Greens')
```

### 2) Pairs Plot

```python
import seaborn as sns

sns.pairplot(data=sns.load_dataset('iris'), hue='species')
```

### 3) Violin / Box Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

data    = sns.load_dataset('tips')
f, axes = plt.subplots(1, 2, sharey=True)

sns.violinplot(x='day', y='total_bill', data=data, ax=axes[0])
sns.boxplot(   x='day', y='total_bill', data=data, ax=axes[1])
```

### 4) Joint Plot

```python
import numpy as np
import seaborn as sns

x, y = np.random.randn(2, 3000)
sns.jointplot(x, y, kind='hex', color='g').set_axis_labels('x', 'y')
```

### 5) Timeseries Plot

```python
import seaborn as sns

sns.lineplot(x='timepoint', y='signal', hue='event', data=sns.load_dataset('fmri'))
```

### 6) Scatter Plot

```python
import seaborn as sns

sns.relplot(x='horsepower', y='mpg', hue='origin', size='weight', sizes=(5, 100), data=sns.load_dataset('mpg'))
```

### 7) Stacked Area Chart

```python
import matplotlib.pyplot as plt
 
x = range(1,6)
y = [
     [1, 4, 6, 18, 9], 
     [2, 2, 7, 10, 1], 
     [2, 8, 5, 10, 6]
    ]

plt.stackplot(x, y, labels=['A','B','C'])
plt.legend(loc='upper left')
```

### 9) Line Chart

```python
import numpy as np
import pandas as pd
import seaborn as sns

values = np.random.RandomState().randn(365, 4).cumsum(axis=0)
dates  = pd.date_range(start='2016-01-01', periods=365)

sns.lineplot(data=pd.DataFrame(values, dates, columns=['A', 'B', 'C', 'D']).rolling(7).mean())
```

### 10) Swarm Plot

```python
import pandas as pd
import seaborn as sns

iris = pd.melt(sns.load_dataset('iris'), id_vars='species', var_name='measurement')

sns.swarmplot(x='measurement', y='value', hue='species', data=iris)
```
