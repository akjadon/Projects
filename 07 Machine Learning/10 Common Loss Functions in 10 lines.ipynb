### 1) Mean Squared Error / Quadratic Loss

```python
import numpy as np
import matplotlib.pyplot as plt

def mse(groundtruth, predictions):
  return np.sum((groundtruth - predictions)**2)

groundtruth = np.repeat(   100, 10000   )
predictions = np.arange(-10000, 10000, 2)
loss        = [mse(groundtruth[i], predictions[i]) for i in range(len(predictions))]

plt.plot(predictions, loss)
plt.xlabel('Prediction')
plt.ylabel('Loss')
plt.title("MSE vs. Prediction")
```

### 2) Mean Absolute Error

```python
import numpy as np
import matplotlib.pyplot as plt

def mae(groundtruth, predictions):
  return np.sum(np.abs(groundtruth - predictions))

groundtruth = np.repeat(   100, 10000   )
predictions = np.arange(-10000, 10000, 2)
loss        = [mae(groundtruth[i], predictions[i]) for i in range(len(predictions))]

plt.plot(predictions, loss)
plt.xlabel('Prediction')
plt.ylabel('Loss')
plt.title("MAE vs. Prediction")
```

### 3) Huber Loss / Smooth Mean Absolute Error

```python
import numpy as np
import matplotlib.pyplot as plt

def huber(groundtruth, predictions, delta):
  return np.sum(np.where(np.abs(groundtruth-predictions) < delta , 0.5*((groundtruth-predictions)**2), delta*np.abs(groundtruth - predictions) - 0.5*(delta**2)))

groundtruth = np.repeat(  0, 1000      )
predictions = np.arange(-10,   10, 0.02)
delta       = [0.1, 1, 10]
loss        = [[huber(groundtruth[i], predictions[i], q) for i in range(len(predictions))] for q in delta]

for i in range(len(delta)):
  plt.plot(predictions, loss[i], label = delta[i])

plt.xlabel('Prediction')
plt.ylabel('Loss')
plt.title('Huber vs. Prediction')
plt.legend(title='delta')
```

### 4) Log-Cosh Loss

```python
import numpy as np
import matplotlib.pyplot as plt

def logcosh(groundtruth, predictions):
  return np.sum(np.log(np.cosh(predictions - groundtruth)))

groundtruth = np.repeat(  0, 1000      )
predictions = np.arange(-10,   10, 0.02)
loss        = [logcosh(groundtruth[i], predictions[i]) for i in range(len(predictions))]

plt.plot(predictions, loss)
plt.xlabel('Prediction')
plt.ylabel('Loss')
plt.title('Log-Cosh vs. Prediction')
```

### 5) Quantile Loss

```python
import numpy as np
import matplotlib.pyplot as plt

def quantile(groundtruth, predictions, quantile):
  return np.sum(np.where(groundtruth >= predictions, quantile*(np.abs(groundtruth-predictions)), (1-quantile)*(np.abs(groundtruth-predictions))))

groundtruth = np.repeat(  0, 1000      )
predictions = np.arange(-10,   10, 0.02)
quantiles   = [0.25, 0.5, 0.75]
loss        = [[quantile(groundtruth[i], predictions[i], q) for i in range(len(predictions))] for q in quantiles]

for i in range(len(quantiles)):
  plt.plot(predictions, loss[i], label = quantiles[i])

plt.xlabel('Prediction')
plt.ylabel('Loss')
plt.title('Quantile vs. Prediction')
plt.legend(title='quantile')
```

### 6) Binary Cross Entropy / Negative Log Likelihood

```python
import numpy as np
import matplotlib.pyplot as plt

def ce(groundtruth, predictions):
  return -np.sum(np.where(groundtruth==1, np.log(predictions), np.log(1-predictions)))

groundtruth = np.repeat(        1,  10000)
predictions = np.arange(0.0001, 1, 0.0001)
loss        = [ce(groundtruth[i], predictions[i]) for i in range(len(predictions))]

plt.plot(predictions, loss)
plt.xlabel('Prediction')
plt.ylabel('Loss')
plt.title('Cross Entropy vs. Prediction')
```

### 7) Focal Loss

```python
import numpy as np
import matplotlib.pyplot as plt

def focal(groundtruth, predictions, gamma):
  return -np.sum(np.where(groundtruth==1, (1-predictions)**gamma*(np.log(predictions)), predictions**gamma*(np.log(1-predictions))))

groundtruth = np.repeat(        1,  10000)
predictions = np.arange(0.0001, 1, 0.0001)
gammas      = [0, 0.5, 1, 2, 5]
loss        = [[focal(groundtruth[i], predictions[i], gamma) for i in range(len(predictions))] for gamma in gammas]

for i in range(len(gammas)):
  plt.plot(predictions, loss[i], label = gammas[i])

plt.xlabel('Prediction')
plt.ylabel('Loss')
plt.title('Focal vs. Prediction')
plt.legend(title='gamma')
```

### 8) Kullback–Leibler Divergence / Relative Entropy

```python
import numpy as np
import matplotlib.pyplot as plt

def kld(groundtruth, predictions):
  return np.sum(predictions*(np.log(predictions) - groundtruth))

groundtruth = np.repeat(     1,  20000)
predictions = np.arange(0.1, 2, 0.0001)
loss        = [kld(groundtruth[i], predictions[i]) for i in range(len(predictions))]

plt.plot(predictions, loss)
plt.xlabel('Prediction')
plt.ylabel('Loss')
plt.title('Kullback–Leibler Divergence vs. Prediction')
```

### 9) Exponential Loss

```python
import numpy as np
import matplotlib.pyplot as plt

def exponential(groundtruth, predictions):
  return np.sum(np.exp(-predictions * groundtruth))

groundtruth = np.repeat(    1,  70000)
predictions = np.arange(-2, 5, 0.0001)
loss        = [exponential(groundtruth[i], predictions[i]) for i in range(len(predictions))]

plt.plot(predictions, loss)
plt.xlabel('Prediction')
plt.ylabel('Loss')
plt.title('Exponential vs. Prediction')
```

### 10) Hinge Loss

```python
import numpy as np
import matplotlib.pyplot as plt

def hinge(groundtruth, predictions):
  return np.sum(np.max((0, (1 - predictions * groundtruth))))

groundtruth = np.repeat(   1,  20000)
predictions = np.arange(0, 2, 0.0001)
loss        = [hinge(groundtruth[i], predictions[i]) for i in range(len(predictions))]

plt.plot(predictions, loss)
plt.xlabel('Prediction')
plt.ylabel('Loss')
plt.title('Hinge vs. Prediction')
```

### 11) Square Loss

```python
import numpy as np
import matplotlib.pyplot as plt

def square(groundtruth, predictions):
  return np.sum((1 - predictions*groundtruth)**2)

groundtruth = np.repeat(   1,  20000)
predictions = np.arange(0, 2, 0.0001)
loss        = [square(groundtruth[i], predictions[i]) for i in range(len(predictions))]

plt.plot(predictions, loss)
plt.xlabel('Prediction')
plt.ylabel('Loss')
plt.title('Square vs. Prediction')
```

### 12) Logistic Loss

```python
import numpy as np
import matplotlib.pyplot as plt

def logistic(groundtruth, predictions):
  return np.sum(np.log(1 + np.exp(-predictions*groundtruth))/np.log(2))

groundtruth = np.repeat(   1,  40000)
predictions = np.arange(0, 4, 0.0001)
loss        = [logistic(groundtruth[i], predictions[i]) for i in range(len(predictions))]

plt.plot(predictions, loss)
plt.xlabel('Prediction')
plt.ylabel('Loss')
plt.title('Logistic vs. Prediction')
```
